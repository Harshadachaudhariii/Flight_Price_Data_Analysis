Description:
The various features of the data are explained below:

1. Airline: The name of the airline company is stored in the airline column. It is a categorical feature having 6 different airlines.
2. source: City from which the flight takes off. It is a catrgorical data having 6 unique cities
3. stops: A categorical feature with 3 distinct values that stores the number of stops b/w the source and destination cities
4. Arrival Time: This is a derived categorical feature created by grouping time intervals into bins. It has six distinct time labels and keep information about the arrival time.
5. Destination: City where the flight will land. It is a categorical feature having 6 unique cities.
6. Duration: A continuous feature that displays the overall amount of time it takes to travel between cities in hours.
7. Price: Target variable stores information of the ticket price.
